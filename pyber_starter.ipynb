```python

%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load 
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Reading the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
#print(city_data)

ride_data = pd.read_csv(ride_data_to_load)
#print(ride_data)

# Combine the data into a single dataset
combined_data = pd.merge(ride_data, city_data, how="left", on=["city", "city"])

# Display the data table for preview
combined_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types;

#creating three different tables based on combined_data: urban, suburban, rural
type_urban = combined_data[combined_data["type"] == "Urban"]
type_suburban = combined_data[combined_data["type"] == "Suburban"]
type_rural = combined_data[combined_data["type"] == "Rural"]

#counting the number of rides based upon the city type;
urban_rides_count = type_urban.groupby(["city"]).count()["ride_id"]
suburban_rides_count = type_suburban.groupby(["city"]).count()["ride_id"]
rural_rides_count = type_rural.groupby(["city"]).count()["ride_id"]

#calculating the average fare based upon city type
urban_average_fare = type_urban.groupby(["city"]).mean()["fare"]
suburban_average_fare = type_suburban.groupby(["city"]).mean()["fare"]
rural_average_fare = type_rural.groupby(["city"]).mean()["fare"]

#calculating the number of drivers based upon city type
driver_count_urban = type_urban.groupby(["city"]).mean()["driver_count"]
driver_count_suburban = type_suburban.groupby(["city"]).mean()["driver_count"]
driver_count_rural = type_rural.groupby(["city"]).mean()["driver_count"]


# Building the scatter plots for each city types;
plt.scatter(urban_rides_count, urban_average_fare,s=10*driver_count_urban,
            marker="o", facecolors="coral", edgecolors="black", alpha=0.75,
            label="Urban")

plt.scatter(suburban_rides_count, suburban_average_fare, s=10*driver_count_suburban,
            marker="o", facecolors="skyblue", edgecolors="black", alpha=0.75,
             label="Suburban")


plt.scatter(rural_rides_count, rural_average_fare, s=10*driver_count_rural,
            marker="o", facecolors="gold", edgecolors="black", alpha=0.7,
            label= "Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")

# Create a legend
legd = plt.legend(fontsize="medium", mode="expanded",
                   numpoints=5, scatterpoints=1,
                   loc="best", title="City Types")
            
legd.legendHandles[0]._sizes = [45]
legd.legendHandles[1]._sizes = [45]
legd.legendHandles[2]._sizes = [45]
                                    
# Incorporate a text label regarding circle size
#lt.text(30, 40, "Note:\n Circle size correlates with driver count per city")
plt.text(42, 42, "Note:\n Circle size correlates with driver count per city")
#set a grid on plot
plt.grid()

# Save Figure
plt.savefig("../Images/pyber bubble plot.png")
# or we can do, os.path("C,"","") by importing os on the top and eliminating "\" with "," from the file path.
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Percent of total fare by city type
type_percent = combined_data.groupby(["type"]).sum()["fare"] / combined_data["fare"].sum() * 100

# Build Pie Chart
plt.pie(type_percent,
        labels = ["Rural", "Suburban", "Urban"],
        colors = ["yellow", "lightskyblue", "lightcoral"],
        explode=[0, 0, 0], 
        autopct="%1.1f%%",
        shadow=True,
        startangle=130)

plt.title("% of Total Fares by City Type")

plt.axis("equal")

# Save Figure
plt.savefig("../Images/pyber total fare.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Percent of total rides by city type
total_rides_percent = combined_data.groupby(["type"]).count()["ride_id"] / combined_data["ride_id"].count() * 100

# Build Pie Chart
plt.pie(total_rides_percent,
        labels = ["Rural", "Suburban", "Urban"],
        colors =  ["yellow", "lightskyblue", "lightcoral"],
        explode = [0, 0, 0],
        autopct="%1.1f%%",
        shadow=True,
        startangle=140)
        
plt.title("% of Total Rides by City Type")

plt.axis("equal")

# Save Figure
plt.savefig("../Images/Pyber total ride.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Percent of total drivers by city type
driver_percent = 100 * city_data.groupby(["type"]).sum()["driver_count"] / city_data["driver_count"].sum() 

# Build Pie Charts
plt.pie(driver_percent,
        labels = ["Rural", "Suburban", "Urban"],
        colors = ["yellow", "lightskyblue", "lightcoral"],
        explode = [0, 0, 0],
        autopct="%1.1f%%",
        shadow=True,
        startangle=140)

plt.title("% of Total Drivers by City Type")

plt.axis("equal")


# Save Figure
plt.savefig("../Images/Pyber total drivers.png")
```

```python
# Show Figure
plt.show()
```
